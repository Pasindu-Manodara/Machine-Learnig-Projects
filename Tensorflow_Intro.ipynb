```python
import numpy as np
import tensorflow as tf
```

```python
x = tf.Variable(3.0)
```

```python
with tf.GradientTape() as tape:
  y = x * x
```

```python
y
```

```python
tape.gradient(y, x)
```

```python
x = tf.Variable([
  [1.0, -1.0, 2.0, 4.0],
  [-2.0, 3.0, -5.0, 9.0],
  [4.0, 2.0, 3.0, 9.0],
  [1.0, 5.0, 6.0, 3.0],

])
t = 30.0
```

```python
with tf.GradientTape() as tape:
  a = tf.nn.max_pool2d(tf.reshape(x, (1, 4, 4, 1)), ksize=2, strides=2, padding="VALID")
  b = tf.reduce_sum(a)
  error = 0.5 * (b - t) ** 2
```

```python
a
```

```python
error
```

```python
tape.gradient(error, [b, x])
```
